```python
# ================================
# 2. Imports
# ================================
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
# ================================
# 3. Dataset paths
# ================================
base_dir = "/content/chest_xray"

train_dir = os.path.join(base_dir, "train")
val_dir   = os.path.join(base_dir, "val")
test_dir  = os.path.join(base_dir, "test")

print("Train path exists:", os.path.exists(train_dir))
print("Val path exists:", os.path.exists(val_dir))
print("Test path exists:", os.path.exists(test_dir))
```

```python
# ================================
# 4. Data generators
# ================================
IMG_SIZE = 150
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255
)

val_test_datagen = ImageDataGenerator(
    rescale=1./255
)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    color_mode='grayscale',
    shuffle=True
)

val_generator = val_test_datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    color_mode='grayscale',
    shuffle=False
)

test_generator = val_test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    color_mode='grayscale',
    shuffle=False
)

print("Class indices:", train_generator.class_indices)
```

```python
# ================================
# 5. Simple CNN model
# ================================
model = models.Sequential([
    layers.Input(shape=(IMG_SIZE, IMG_SIZE, 1)),

    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# ================================
# 6. Train
# ================================
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=3,
        restore_best_weights=True
    ),
    ModelCheckpoint(
        "basic_cnn_pneumonia.keras",
        save_best_only=True,
        monitor='val_loss'
    )
]

EPOCHS = 10

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    callbacks=callbacks
)
```

```python
# ================================
# 7. Evaluate
# ================================
test_loss, test_acc = model.evaluate(test_generator)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
# ================================
# 8. Plot training history
# ================================
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Train Accuracy')
plt.plot(epochs_range, val_acc, label='Val Accuracy')
plt.legend()
plt.title('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Train Loss')
plt.plot(epochs_range, val_loss, label='Val Loss')
plt.legend()
plt.title('Loss')

plt.show()
```

```python
# ================================
# 9. Sample predictions
# ================================
import random
from tensorflow.keras.utils import load_img, img_to_array

class_names = {v: k for k, v in train_generator.class_indices.items()}

test_normal_dir = os.path.join(test_dir, "NORMAL")
test_pneu_dir = os.path.join(test_dir, "PNEUMONIA")

sample_paths = []
sample_paths += [os.path.join(test_normal_dir, f) for f in random.sample(os.listdir(test_normal_dir), 3)]
sample_paths += [os.path.join(test_pneu_dir, f) for f in random.sample(os.listdir(test_pneu_dir), 3)]

plt.figure(figsize=(15, 8))

for i, img_path in enumerate(sample_paths, 1):
    img = load_img(img_path, color_mode='grayscale', target_size=(IMG_SIZE, IMG_SIZE))
    img_array = img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    pred_prob = model.predict(img_array, verbose=0)[0][0]
    pred_class = "PNEUMONIA" if pred_prob >= 0.5 else "NORMAL"

    plt.subplot(2, 3, i)
    plt.imshow(img, cmap='gray')
    plt.title(f"Pred: {pred_class}\nProb: {pred_prob:.3f}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
```

```python
base_dir = "/content/chest_xray"

train_dir = base_dir + "/train"
val_dir = base_dir + "/val"
test_dir = base_dir + "/test"

IMG_SIZE = 150
BATCH_SIZE = 32
```

```python
def add_laplace_noise(image, epsilon=1.0):

    sensitivity = 1.0
    scale = sensitivity / epsilon

    noise = np.random.laplace(loc=0.0, scale=scale, size=image.shape)

    noisy_image = image + noise

    noisy_image = np.clip(noisy_image, 0, 1)

    return noisy_image
```

```python
train_datagen_normal = ImageDataGenerator(rescale=1./255)

train_generator_normal = train_datagen_normal.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="binary"
)
```

```python
train_datagen_laplace = ImageDataGenerator(
    rescale=1./255,
    preprocessing_function=lambda img: add_laplace_noise(img)
)

train_generator_laplace = train_datagen_laplace.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="binary"
)
```

```python
val_datagen = ImageDataGenerator(rescale=1./255)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="binary"
)
```

```python
def build_model():

    model = models.Sequential([
        layers.Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE,IMG_SIZE,1)),
        layers.MaxPooling2D(2,2),

        layers.Conv2D(64,(3,3),activation='relu'),
        layers.MaxPooling2D(2,2),

        layers.Conv2D(128,(3,3),activation='relu'),
        layers.MaxPooling2D(2,2),

        layers.Flatten(),
        layers.Dense(128,activation='relu'),
        layers.Dropout(0.3),

        layers.Dense(1,activation='sigmoid')
    ])

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
tf.keras.backend.clear_session()

model_normal = build_model()

history_normal = model_normal.fit(
    train_generator_normal,
    validation_data=val_generator,
    epochs=10
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    color_mode="grayscale",
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=False
)
```

```python
test_loss_normal, test_acc_normal = model_normal.evaluate(test_generator)

print("Normal test accuracy:", test_acc_normal)
```

```python
tf.keras.backend.clear_session()

model_laplace = build_model()

history_laplace = model_laplace.fit(
    train_generator_laplace,
    validation_data=val_generator,
    epochs=10
)
```

```python
test_loss_laplace, test_acc_laplace = model_laplace.evaluate(test_generator)

print("Laplace test accuracy:", test_acc_laplace)
```

```python
epochs = range(1, len(history_normal.history['accuracy']) + 1)

plt.figure(figsize=(8,5))

plt.plot(epochs, history_normal.history['accuracy'], label="Train Normal")
plt.plot(epochs, history_laplace.history['accuracy'], label="Train Laplace")

plt.plot(epochs, history_normal.history['val_accuracy'], label="Val Normal")
plt.plot(epochs, history_laplace.history['val_accuracy'], label="Val Laplace")

plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Normal vs Laplace Training")
plt.legend()

plt.show()
```
